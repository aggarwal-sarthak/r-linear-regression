```r
library(readr)
library(ggplot2)
```

```r
data<-read.csv("http://bit.ly/w-data")
model<- lm(Scores~Hours,data)
model
```

```r
ggplot(data,aes(x=Hours,y=Scores))+geom_point()
```

```r
plot(data)
abline(model, col="yellow")
```

```r
a<-data.frame(Hours=9.25)
result<-predict(model,a)
print(result)
```
